# Fraud Detection Analysis with Python

This notebook analyzes financial transaction data to identify fraud patterns. The focus is on data exploration, fraud distribution, transaction types, and simple business insights.

## Business Questions

1. How many transactions are included in the dataset?
2. What percentage of transactions are fraudulent?
3. Which transaction types are most common?
4. Which transaction types are most associated with fraud?
5. What is the total and average amount involved in fraudulent transactions?
6. How do transaction amounts differ between fraud and non-fraud transactions?
7. How does fraud activity change over time?
8. How often does the system flag transactions that are actually fraudulent?
9. Are there missing values that should be reviewed before deeper analysis?

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
# Note: This project uses a sample dataset for GitHub because the full dataset is too large.

df = pd.read_csv('../data/fraud_transactions_sample.csv')
df.head()
```

## 1. Dataset Overview

```python
# Check the number of rows and columns
df.shape
```

```python
# Review column names and data types
df.info()
```

```python
# Basic statistics for numeric columns
df.describe()
```

## 2. Missing Values Check

```python
# Check missing values by column
df.isnull().sum()
```

## 3. Overall Fraud Rate

```python
total_transactions = len(df)
total_fraud = df['isFraud'].sum()
fraud_rate = total_fraud / total_transactions * 100

print(f'Total transactions: {total_transactions:,}')
print(f'Fraudulent transactions: {total_fraud:,}')
print(f'Fraud rate: {fraud_rate:.2f}%')
```

## 4. Transaction Type Distribution

```python
transaction_type_counts = df['type'].value_counts()
transaction_type_counts
```

```python
transaction_type_counts.plot(kind='bar')
plt.title('Transaction Count by Type')
plt.xlabel('Transaction Type')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. Fraud by Transaction Type

```python
fraud_by_type = df.groupby('type')['isFraud'].agg(['count', 'sum'])
fraud_by_type = fraud_by_type.rename(columns={'count': 'total_transactions', 'sum': 'fraud_transactions'})
fraud_by_type['fraud_rate_pct'] = (fraud_by_type['fraud_transactions'] / fraud_by_type['total_transactions'] * 100).round(2)
fraud_by_type.sort_values('fraud_rate_pct', ascending=False)
```

```python
fraud_by_type['fraud_transactions'].sort_values(ascending=False).plot(kind='bar')
plt.title('Fraudulent Transactions by Type')
plt.xlabel('Transaction Type')
plt.ylabel('Number of Fraudulent Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Fraud Amount Analysis

```python
fraud_transactions = df[df['isFraud'] == 1]

total_fraud_amount = fraud_transactions['amount'].sum()
avg_fraud_amount = fraud_transactions['amount'].mean()
median_fraud_amount = fraud_transactions['amount'].median()

print(f'Total fraud amount: ${total_fraud_amount:,.2f}')
print(f'Average fraud amount: ${avg_fraud_amount:,.2f}')
print(f'Median fraud amount: ${median_fraud_amount:,.2f}')
```

## 7. Fraud vs Non-Fraud Amount Comparison

```python
amount_summary = df.groupby('isFraud')['amount'].agg(['count', 'mean', 'median', 'min', 'max'])
amount_summary
```

```python
df.boxplot(column='amount', by='isFraud')
plt.title('Transaction Amount by Fraud Status')
plt.suptitle('')
plt.xlabel('Fraud Status (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Transaction Amount')
plt.tight_layout()
plt.show()
```

## 8. Fraud Over Time

```python
fraud_over_time = df[df['isFraud'] == 1].groupby('step').size()
fraud_over_time.head()
```

```python
fraud_over_time.plot(kind='line')
plt.title('Fraudulent Transactions Over Time')
plt.xlabel('Step')
plt.ylabel('Number of Fraudulent Transactions')
plt.tight_layout()
plt.show()
```

## 9. Flagged Fraud Review

```python
flagged_review = df.groupby(['isFraud', 'isFlaggedFraud']).size().reset_index(name='transaction_count')
flagged_review
```

## 10. Key Takeaways

- Fraud transactions are rare compared with normal transactions.
- Fraud is concentrated in specific transaction types.
- Fraudulent transaction amounts tend to be larger than normal transaction amounts.
- The system flag does not capture all fraud cases, so further review or better detection rules may be needed.
- This analysis can be expanded later with feature engineering and machine learning, but the first priority is understanding the fraud patterns clearly.
